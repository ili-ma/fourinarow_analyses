```python
import numpy as np
import matplotlib.pyplot as plt
import re
direc = 'C:/Users/Bas/Google Drive/Bas Games/'
```

```python
def get_params(n):
    i = re.search("\d", n).start()
    return np.loadtxt(direc + 'Analysis/Params/params_' + n[:i] + '_final_short.txt')[int(n[i:])]
```

```python
with open(direc + 'Analysis/Ratings/ratings_gen_eye_learn_tai.txt') as file:
    data = [[line.split()[1][6:],int(line.split()[2])] 
            for line in file if line.split()[1].startswith('model_')]
```

```python
params=[get_params(min(data, key=lambda x:abs(x[1]-e))[0]) for e in np.linspace(-200,400,200)]
```

```python
np.savetxt(direc + 'params_peak.txt',params,fmt='%.05f',delimiter=',',newline='},\n{')
```

```python
plt.plot(np.array(data)[:,1])
plt.show()

plt.plot(np.array(np.array(params)[:,2]),'.')
plt.show()
```
