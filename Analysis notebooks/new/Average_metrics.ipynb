```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import pandas as pd
from matplotlib import rcParams,font_manager
from matplotlib.colors import ListedColormap

from fourinarowfunctions import *

datadir = "../data/"

#Average over folds and create StudyID variable.
allMetrics = pd.read_csv(datadir + "params_with_metrics.csv")
age = pd.read_csv(datadir + "SubjectAge.csv")
allMetrics['StudyID'] = allMetrics['subject']
allMetrics["StudyID"]= allMetrics["StudyID"].str.split(":", n = 1, expand = True)
allMetrics.head()
```

```python
# merge dataframes, average across folds and drop fold column
av_metrics = pd.merge(allMetrics, age, on="StudyID", how="left")
av_metrics = av_metrics.groupby('subject').mean()
av_metrics = av_metrics.drop(columns=['fold'])
av_metrics.head()
```

```python
av_metrics.to_csv('av_Metrics.csv')
```
