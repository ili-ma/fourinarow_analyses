```python
from fourinarowfunctions import *
import statistics

datadir = '../data/'
filename = datadir + 'trialdata.csv'
data_dict = load_data(filename, verbose = False)
parsed_dict = {}
for username, data in data_dict.items():
    try:
        parsed_dict[username] = get_parsed_data(data, username)
    except AssertionError as e:
        continue
usernames = sorted(parsed_dict.keys())
usernames.remove("A1153:R_1CwnNA5xXOlHq5l")
usernames.remove("866:R_31YEgt5MqfcfZMC")
print(f"{len(usernames)} users with complete data")
```

```python
def get_repeats(username, user_data):
    seen_states = set()
    unique = 0
    repeated = 0
    empty = 0
    for trialnr, trial in enumerate(user_data):
        for statenr, state in enumerate(trial):
            state_key = f"{state['user_color']}-{state['bp']}-{state['wp']}"
            if state['bp'] == state['wp']:
                empty += 1
            elif state_key in seen_states:
                repeated += 1
            else:
                unique += 1
            seen_states.add(state_key)
    perc_unique = 100.0 * unique / (unique + repeated + empty)
    print(f"{username:25} unique: {unique:3} repeated: {repeated:3} empty: {empty:2}  % unique: {perc_unique}")
    return perc_unique

percentages = []
for username in usernames:
    percentages.append(get_repeats(username, parsed_dict[username]))

print(f"\nAverage: {statistics.mean(percentages)}\nStDev: {statistics.stdev(percentages)}")
```
